## Feature Engineering

```python
import pandas as pd
```

```python
df = pd.read_excel("uploads/enhanced_employee_database.xlsx")
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace(r'[^0-9a-zA-Z]', "", regex = True)
df['Phone_Number'] = df['Phone_Number'].fillna("")
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[-10:] if isinstance(x,str) and len(x) > 10 else x )

df
```

```python
# Label coding
# df['Performance_Rating']= df['Performance_Rating'].astype('category').cat.codes
# df['Performance_Rating']

# OR

rating_order = ['Poor','Average','Good','Excellent']
df['Performance_Rating'] = (
    df['Performance_Rating'] .astype(
        (pd.CategoricalDtype(categories = rating_order, ordered = True))
    ).cat.codes
)

df
```
